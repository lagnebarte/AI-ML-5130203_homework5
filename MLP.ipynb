```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
dataset = pd.read_csv("data/weather_forecast_data (2).csv", header = 0)
dataset.head()
```

```python
# split independent and dependent variables
x = dataset.iloc[:200, [1,3]].values
y = dataset.iloc[:200, -1].values
```

```python
x[:5]
```

```python
y = np.where(y == 'no rain', 1, 0)
y
```

```python
def explore_axis(x, y):
    class_norain = x[y == 1]
    class_rain = x[y == 0]

    plt.figure(figsize=(8, 6))
    plt.scatter(class_norain[:, 0], class_norain[:, 1], color='blue', label='no rain', alpha=0.7)
    plt.scatter(class_rain[:, 0], class_rain[:, 1], color='orange', label='rain', alpha=0.7)

    plt.xlabel("Humidity")
    plt.ylabel("Cloud_Cover")
    plt.title("Classification Visualization: No Rain vs Rain")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
explore_axis(x, y)
```

```python
import mlp_test
from scripts.mlp import MLPAlgorithm
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
print(f"Train size: {x_train.shape}, Test size: {x_test.shape}")
print(f"Train labels: {y_train[:10]}, Test labels: {y_test[:10]}")
```

```python
input_length = x_train.shape[1] 
hidden_length = 10 
output_length = 1

#    
model = MLPAlgorithm(eta=0.01, max_epochs=100, threshold=1e-4)
model.build_architecture(input_length, hidden_length, output_length)

#  
model.fit(x_train, y_train)
```

```python
model.test(x_test, y_test)
print(f"Accuracy = {model.accuracy}%")
```

```python
from scripts.Eval import EvalClassifier
```

```python
from sklearn.metrics import classification_report

y_pred_test = model.test(x_test, y_test)
y_pred_train = model.test(x_train, y_train)

evaluator = EvalClassifier(model)

metrics_test = evaluator.evaluate_test(y_test, y_pred_test, classes=[0, 1])
print("Test:")
print(metrics_test)

metrics_train = evaluator.evaluate_train(y_train, y_pred_train, classes=[0, 1])
print("Train:")
print(metrics_train)

print("\nClassification Report (Test):")
print(classification_report(y_test, y_pred_test, target_names=["rain", "no rain"]))

print("\nClassification Report (Train):")
print(classification_report(y_train, y_pred_train, target_names=["rain", "no rain"]))
```
